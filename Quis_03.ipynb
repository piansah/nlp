```python
import pandas as pd
import numpy as np
import re
```

```python
bahasa = pd.read_csv('/content/2023_nlp_kuis_train.csv')
bahasa
```

```python
bahasa_ah = bahasa['ah']
bahasa_ah
```

```python
bahasa_uh = bahasa['uh']
bahasa_uh
```

```python
soal_dua = pd.read_csv('/content/soal_2.csv')
soal_dua
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

token_ah = Tokenizer()
token_uh = Tokenizer()
token_ah.fit_on_texts(bahasa_ah)
token_uh.fit_on_texts(bahasa_uh)
```

```python
v_bahasa_ah = token_ah.texts_to_sequences(bahasa_ah)
v_bahasa_ah
```

```python
v_bahasa_uh = token_uh.texts_to_sequences(bahasa_uh)
v_bahasa_uh
```

```python
total_kata_bahasa_ah = len(token_ah.word_index)
total_kata_bahasa_uh = len(token_uh.word_index)
```

```python
v_bahasa_ah = pad_sequences(v_bahasa_ah, maxlen = 57, padding = 'post')
v_bahasa_ah
```

```python
v_bahasa_uh = pad_sequences(v_bahasa_uh, maxlen = 57, padding = 'post')
v_bahasa_uh
```

```python
v_bahasa_ah.shape
v_bahasa_uh.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, GRU, TimeDistributed, LSTM, Embedding, Bidirectional

model = Sequential()
model.add(Embedding(total_kata_bahasa_uh+1, 64, input_shape = v_bahasa_uh.shape[1:], input_length = v_bahasa_uh.shape[1]))
model.add(Bidirectional(GRU(128, return_sequences = True)))
model.add(TimeDistributed(Dense(64, activation = 'relu')))
model.add(TimeDistributed(Dense(32, activation = 'relu')))
model.add(TimeDistributed(Dense(16, activation = 'relu')))
model.add(TimeDistributed(Dense(8, activation = 'relu')))
model.add(TimeDistributed(Dense(total_kata_bahasa_ah+1, activation = 'softmax')))
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics =['accuracy'])
model.summary()
```

```python
model.fit(
    v_bahasa_uh,
    v_bahasa_ah,
    batch_size = 2,
    epochs = 10,
    validation_split = 0.2
)
```
