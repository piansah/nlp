```python
daftar = ['daftar', 'pendaftaran', 'cara daftar']
```

```python
bayar = ['bayar', 'pembayaran', 'cara bayar']
```

```python
r_daftar = 'cara daftarnya lewat link ini ...'
```

```python
r_bayar = 'kirim ke BNI ... untuk membayar'
```

```python
pertanyaan = input()
```

```python
status_bayar = False
status_daftar = False
```

```python
for kata_kunci in daftar:
  if kata_kunci in pertanyaan:
    status_daftar = True
    break
for kata_kunci in bayar:
  if kata_kunci in pertanyaan:
    status_bayar = True
```

```python
if status_daftar == True:
  print(r_daftar)
elif status_bayar == True:
  print(r_bayar)
else:
  print('maaf, apa tadi pertanyaannya? saya tidak paham')
```
